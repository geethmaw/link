```python
from statsmodels.formula.api import ols
import pandas
from scipy.stats import spearmanr
```

```python
### regress variables on parameter values. 
### Parameters are randomly distributed so there is covariance between parameter values.
### correlations are one at a time and a variable may depend simultaneously on many parameters.
def get_corr_pp(xx,param_name,params):
    model=[None]*len(param_name)
    rsq=zeros((1,len(param_name)))
    zzo=[]; nmout=[]
    xxo=zeros(mm.shape[0])*NaN
    #yyo=xxo*NaN
    n=1
    tick=0
    for i in range(len(param_name)):
        if param_name[i] !='Sample_nmb':
            params[param_name[i]]
            zz=params[param_name[i]].data*1.
            if param_name[i]=='micro_mg_dcs':
                zz[zz>1000.*1e-6]=NaN
            xx1=xx[0:len(zz)]; ind=isnan(xx1*zz)==False
            rsq[0,i]=spearmanr(zz[ind],xx1[ind])[0]
            #rsq[2,i]=spearmanr(zz[ind2],xxn)[0]
        tick=tick+1
    return rsq
```

```python
rsq={}
rsq['dwlpdnd']=get_corr_pp(dd,param_name,params)
rsq['slope']=get_corr_pp(mm[:,0],param_name,params)
rsq['inter']=get_corr_pp(iinter[:,0],param_name,params)
rsq['dadlwp']=get_corr_pp(dadlwp,param_name,params)
```

```python
### plot the correlation between diferent variables and the parameters going into the PPE

plt.figure(figsize=(10,5))
plt.bar(arange(len(rsq['dwlpdnd'][0])),rsq['dwlpdnd'][0],tick_label=param_name)
plt.xticks(rotation = 90)
```
